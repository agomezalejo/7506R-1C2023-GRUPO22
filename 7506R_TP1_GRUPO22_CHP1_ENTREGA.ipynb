## Cálculo de las medidas de resumen de las variables cuantitativas 

```python
variables_cuantitativas = [
    'lead_time',
    'arrival_date_year',
    'arrival_date_week_number',
    'arrival_date_day_of_month',
    'stays_in_weekend_nights',
    'stays_in_week_nights',
    'adults',
    'children',
    'babies',
    'previous_cancellations',
    'previous_bookings_not_canceled',
    'days_in_waiting_list',
    'required_car_parking_spaces',
    'total_of_special_requests',
    'reservation_status_date']

df_describe = df[variables_cuantitativas].describe().drop('count', axis=0).transpose()

df_moda = df[variables_cuantitativas].mode().transpose()

df_media = df[variables_cuantitativas].median().transpose()

df_describe.assign(mode=df_moda, median=df_media)
```
